```python
import requests
import time

def fetch_flights(iata_code):

    url = f"https://aerodatabox.p.rapidapi.com/flights/airports/iata/{iata_code}"

    querystring = {
        "offsetMinutes":"-120",
        "durationMinutes":"720",
        "withLeg":"true",
        "direction":"Both",
        "withCancelled":"true",
        "withCodeshared":"true",
        "withCargo":"true",
        "withPrivate":"true",
        "withLocation":"false"
    }

    response = requests.get(
        url,
        headers=headers,
        params=querystring
    )

    time.sleep(2)

    return response.json()
```

```python
import time
import pandas as pd

airport_codes = [
    "DEL", "BLR", "BOM", "MAA", "HYD",
    "CCU", "COK", "AMD", "GOI", "IXA"
]

flight_rows = []

for airport in airport_codes:

    print("Loading", airport)

    try:
        flight_data = fetch_flights(airport)

        # ------------------
        # DEPARTURES
        # ------------------
        for flight in flight_data.get("departures", []):

            flight_rows.append({

                "flight_number":
                    flight.get("number"),

                "aircraft_registration":
                    flight.get("aircraft", {}).get("reg"),

                "aircraft_model":
                    flight.get("aircraft", {}).get("model"),

                "airline":
                    flight.get("airline", {}).get("name"),

                "airline_iata":
                    flight.get("airline", {}).get("iata"),

                "status":
                    flight.get("status"),

                "origin_airport":
                    airport,

                "destination_airport":
                    flight.get("arrival", {})
                          .get("airport", {})
                          .get("iata"),

                "departure_time":
                    flight.get("departure", {})
                          .get("scheduledTime", {})
                          .get("local"),

                "arrival_time":
                    flight.get("arrival", {})
                          .get("scheduledTime", {})
                          .get("local")
            })

        # ------------------
        # ARRIVALS
        # ------------------
        for flight in flight_data.get("arrivals", []):

            flight_rows.append({

                "flight_number":
                    flight.get("number"),

                "aircraft_registration":
                    flight.get("aircraft", {}).get("reg"),

                "aircraft_model":
                    flight.get("aircraft", {}).get("model"),

                "airline":
                    flight.get("airline", {}).get("name"),

                "airline_iata":
                    flight.get("airline", {}).get("iata"),

                "status":
                    flight.get("status"),

                "origin_airport":
                    flight.get("departure", {})
                          .get("airport", {})
                          .get("iata"),

                "destination_airport":
                    airport,

                "departure_time":
                    flight.get("departure", {})
                          .get("scheduledTime", {})
                          .get("local"),

                "arrival_time":
                    flight.get("arrival", {})
                          .get("scheduledTime", {})
                          .get("local")
            })

        print(
            airport,
            "Departures:", len(flight_data.get("departures", [])),
            "Arrivals:", len(flight_data.get("arrivals", []))
        )

        time.sleep(2)

    except Exception as e:
        print(f"Error loading {airport}: {e}")

# ------------------
# CREATE DATAFRAME
# ------------------

flights_df = pd.DataFrame(flight_rows)

# Optional flight_id
flights_df.insert(
    0,
    "flight_id",
    range(1, len(flights_df) + 1)
)

print("\nShape:", flights_df.shape)
print("\nColumns:")
print(flights_df.columns.tolist())

flights_df.head()
```

1. 1)Show the total number of flights for each aircraft model, listing the model and its count.

```python
query = """
SELECT
    aircraft_model,
    COUNT(*) AS total_flights
FROM flights
GROUP BY aircraft_model
ORDER BY total_flights DESC;
"""

pd.read_sql_query(query, conn)
```

2 2)List all aircraft (registration, model) that have been assigned to more than 5 flights.

```python
query = """
SELECT
    aircraft_model,
    COUNT(*) AS flight_count
FROM flights
GROUP BY aircraft_model
HAVING COUNT(*) > 5
ORDER BY flight_count DESC;
"""

pd.read_sql_query(query, conn)
```

3 For each airport, display its name and the number of outbound flights, but only for airports with more than 5 flights.

```python
query = """
SELECT
    a.name,
    COUNT(*) AS outbound_flights
FROM flights f
JOIN airport a
    ON f.origin_airport = a.iata_code
GROUP BY a.name
HAVING COUNT(*) > 5
ORDER BY outbound_flights DESC;
"""

pd.read_sql_query(query, conn)
```

4 4)Find the top 3 destination airports (name, city) by number of arriving flights, sorted by count descending.

```python
query = """
SELECT
    a.name,
    a.city,
    COUNT(*) AS arrivals
FROM flights f
JOIN airport a
    ON f.destination_airport = a.iata_code
GROUP BY a.name, a.city
ORDER BY arrivals DESC
LIMIT 3;
"""

pd.read_sql_query(query, conn)
```

5 5)Show for each flight: number, origin, destination, and a label 'Domestic' or 'International' using CASE WHEN on country match.

```python
query = """
SELECT
    f.flight_number,
    f.origin_airport,
    f.destination_airport,

    CASE
        WHEN ao.country = ad.country
            THEN 'Domestic'
        ELSE 'International'
    END AS flight_type

FROM flights f

LEFT JOIN airport ao
    ON f.origin_airport = ao.iata_code

LEFT JOIN airport ad
    ON f.destination_airport = ad.iata_code;
"""

pd.read_sql_query(query, conn)
```

6 6)Show the 5 most recent arrivals at “DEL” airport including flight number, aircraft, departure airport name, and arrival time, ordered by latest arrival.

```python
query = """
SELECT
    flight_number,
    aircraft_model,
    origin_airport,
    arrival_time
FROM flights
WHERE destination_airport = 'DEL'
ORDER BY arrival_time DESC
LIMIT 5;
"""

pd.read_sql_query(query, conn)
```

7 7) Find all airports with no arriving flights (never used as a destination in flights table).

```python
query = """
SELECT
    a.name,
    a.city,
    a.iata_code
FROM airport a
LEFT JOIN flights f
    ON a.iata_code = f.destination_airport
WHERE f.destination_airport IS NULL;
"""

pd.read_sql_query(query, conn)
```

8 8)For each airline, count the number of flights by status (e.g., 'On Time', 'Delayed', 'Cancelled') using CASE WHEN.

```python
query = """
SELECT
    airline,

    SUM(CASE WHEN status='On Time' THEN 1 ELSE 0 END) AS on_time,

    SUM(CASE WHEN status='Delayed' THEN 1 ELSE 0 END) AS delayed,

    SUM(CASE WHEN status='Canceled' THEN 1 ELSE 0 END) AS canceled,

    COUNT(*) AS total_flights

FROM flights
GROUP BY airline
ORDER BY total_flights DESC;
"""

pd.read_sql_query(query, conn)
```

9 9)Show all cancelled flights, with aircraft and both airports, ordered by departure time descending.

```python
query = """
SELECT
    flight_number,
    aircraft_model,
    origin_airport,
    destination_airport,
    departure_time
FROM flights
WHERE status='Canceled'
ORDER BY departure_time DESC;
"""

pd.read_sql_query(query, conn)
```

 10 List all city pairs (origin-destination) that have more than 2 different aircraft models operating flights between them.

```python
query = """
SELECT
    ao.city AS origin_city,
    ad.city AS destination_city,
    COUNT(DISTINCT f.aircraft_model) AS aircraft_models

FROM flights f

JOIN airport ao
    ON f.origin_airport = ao.iata_code

JOIN airport ad
    ON f.destination_airport = ad.iata_code

GROUP BY
    ao.city,
    ad.city

HAVING COUNT(DISTINCT f.aircraft_model) > 2

ORDER BY aircraft_models DESC;
"""

pd.read_sql_query(query, conn)
```

11) For each destination airport, compute the % of delayed flights (status='Delayed') among all arrivals, sorted by highest percentage.

```python
query = """
SELECT
    destination_airport,

    ROUND(
        100.0 *
        SUM(CASE WHEN status='Delayed' THEN 1 ELSE 0 END)
        / COUNT(*),
        2
    ) AS delayed_percentage

FROM flights

GROUP BY destination_airport

ORDER BY delayed_percentage DESC;
"""

pd.read_sql_query(query, conn)
```
